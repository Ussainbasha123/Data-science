```

import matplotlib.pyplot as plt
```

```

d.groupby("gender")[['math score', 'reading score',
       'writing score']].mean().plot(kind="bar")
plt.title("gender vs score")
plt.xlabel("gender")
plt.ylabel("score")
plt.show()
```

```
c=["one piece","bleach","black clover"]
s=[1155,387,180]
plt.bar(c,s,color=["red","gray","blue"])
plt.title("animie")
plt.xlabel("Best animies")
plt.ylabel("no of episodes")
plt.show()
```

```

marks=[89,93,45,67,92,97,54,53,52,100]
plt.hist(marks)
plt.show()
```

```

plt.hist(d["math score"],bins=20)
plt.title("distribution of math score")
plt.xlabel("score")
plt.ylabel("frequency")
plt.show()
```

```

#scatter plot
plt.scatter(d["math score"],d["writing score"])
plt.title("math score vs writing score")
plt.xlabel("math score")
plt.ylabel("writing score")
plt.show()
```

```

avg=d[['math score','writing score','reading score']].mean()
avg
```

```
x=avg.index
y=avg.values
print(x)
print(y)
```

```

plt.plot(x,y)
plt.title("maths vs reading")
plt.xlabel("maths")
plt.ylabel("reading")
plt.show()
```
