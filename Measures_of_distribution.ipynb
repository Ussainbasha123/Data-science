```python
import numpy as np
```

```python
normal_data=np.random.normal(loc=50,scale=5,size=1000)
normal_data
```

```python
a=np.array([100,150,200,300,400])
a_data=np.concatenate([normal_data,a])
a_data
```

```python
nm=np.mean(normal_data)
am=np.mean(a_data)
print(nm)
print(am)
```
